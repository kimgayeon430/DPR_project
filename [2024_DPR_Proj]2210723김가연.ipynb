# **CNN 모델 생성**
 레이어를 3층으로 주었고, 각 레이어는 가장 먼저 특정 특징을 활성화하는 ***컨벌루션(합성곱) 계층(Conv2D)***이 있고, 이 계층에서 활성화함수를 ***ReLU***로 설정하여 더 빠르고 효과적으로 훈련이 이루어지도록 하였다. 이 계층에서는 입력데이터가 가진 특징을 대략적으로 추출하여 학습을 진행할 수 있다.

 그 다음 계층에서는 입력 데이터를 평균 0, 분산 1로 정규화하여 학습이 잘 이루어지도록 하는 방법인 ***배치정규화(Batch Normalization)***을 사용하였다.

 그 후, 풀링 계층을 통과하게 되는데, 이는 ***Maxpooling***을 사용하여 앞서 도출된 이미지 특징을 가시 한 번 축소하는 풀링(pooling)과정이다. 해당 모델에서는 정해진 구역 안에서 최댓값을 뽑아내는 맥스풀링을 사용하였다.

 마지막으로 과적합을 방지하기 위해 은닉층에 배치된 노드 중 일부를 임의로 꺼주는 ***Dropout*** 과정이 있다. 해당 모델에서는 0.25로 드롭아웃을 적용하여 25%의 노드를 임의로 끈 후 학습하게 된다.

 이렇게 3개의 레이어를 거치게 되면 마지막으로 완전 연결 계층(Fully connected layer)을 거친다. 해당 모델에서는 다시 활성화함수가 있는 층에서 사용하기 위해 2차원 배열을 1차원으로 바꾸어주는 Flatten()함수를 통과하게 된다.

```python
from keras.models import Sequential
from keras.layers import *
```

```python
# 레이어 1
model = Sequential()
model.add(Conv2D(32, (3,3), activation="relu", input_shape=(IMAGE_WIDTH, IMAGE_HEIGHT, IMAGE_CHANNEL)))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))

# 레이어 2
model.add(Conv2D(64, (3,3), activation="relu"))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))

# 레이어3
model.add(Conv2D(128, (3,3), activation="relu"))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))

# Fully Connected
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))

model.add(Dense(2,activation="softmax"))
```

```python
# 모델 실행 옵션
model.compile(loss="categorical_crossentropy", optimizer="rmsprop", metrics=['accuracy'])
model.summary()
```

learning late가 업데이트되지 않으면 학습을 중단하는 조기 종료(EarlyStopping) 옵션을 사용하였다.

```python
from keras.callbacks import EarlyStopping, ReduceLROnPlateau

earlystop = EarlyStopping(patience=10)
learning_rate_reduction=ReduceLROnPlateau(
                        monitor= "val_accuracy",
                        patience = 2,
                        factor = 0.5,
                        min_lr=0.0001,
                        verbose=1)

callbacks = [earlystop, learning_rate_reduction]
```

```python
df['category']=df['category'].replace({0:'boy',1:'girl'})
```

test, train 데이터 분리

test size를 0.2로 하여 검증셋을 분리하였다.

```python
from sklearn.model_selection import train_test_split

train_df, validate_df = train_test_split(df , test_size=0.2, random_state= 42)

train_df=train_df.reset_index(drop=True)
validate_df=validate_df.reset_index(drop=True)
```

```python
train_df['category'].value_counts()
```

```python
validate_df['category'].value_counts()
```

```python
from keras.preprocessing.image import *

total_train=train_df.shape[0]
total_validate=validate_df.shape[0]
batch_size=10

# 트레이닝 데이터의 제너레이터 설정
train_datagen=ImageDataGenerator(
        rotation_range=10,
        rescale=1./255,
        shear_range=0.1,
        zoom_range=0.2,
        horizontal_flip=True,
        width_shift_range=0.1,
        height_shift_range=0.1)

train_generator=train_datagen.flow_from_dataframe(
    train_df,
    path+"train",
    x_col = "filename",
    y_col = "category",
    target_size = IMAGE_SIZE,
    class_mode = "categorical",
    batch_size = batch_size )

validate_datagen=ImageDataGenerator(rescale=1./255)

validation_generator=validate_datagen.flow_from_dataframe(
    validate_df,
    path+"train",
    x_col= "filename",
    y_col= "category",
    target_size = IMAGE_SIZE,
    class_mode = "categorical",
    batch_size = batch_size )
```

# **학습**

```python
epochs = 10

history = model.fit_generator(
    train_generator,
    epochs = epochs,
    steps_per_epoch = total_train//batch_size ,
    validation_data=  validation_generator,
    validation_steps = total_validate//batch_size,
    callbacks = callbacks,
)

model.save_weights("model.h5") #모델 저장
```

트레이닝 모델 히스토리 시각화

```python
historyDict=history.history

acc=history.history['accuracy']
val_acc=history.history['val_accuracy']
loss=history.history['loss']
val_loss=history.history['val_loss']

epo = range(1, len(acc)+1)
plt.plot(epo, loss, 'bo', label="Traing loss")
plt.plot(epo, val_loss, 'b', label="Val loss")
plt.xlabel("epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()  #학습이 진행될 때의 loss 그래프 띄움
```

```python
plt.plot(epo, acc, 'ro', label="Traing accuracy")
plt.plot(epo, val_acc, 'r', label="Val accuracy")
plt.xlabel("epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.show()  # 학습이 진행될 때의 정확도 그래프 출력
```
